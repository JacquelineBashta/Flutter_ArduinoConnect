```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv(r"D:\_1_Technicals\DataScience_Bootcamp\Final_project\Flutter_ArduinoConnect\00_DataSets\1_3\t_stand_front_drinksmallcup_10ml_1s_right.csv")
```

```python
df=df.iloc[0:100]
```

```python
df.columns= ["time","rotR_x","rotR_y","rotR_z","acc_x","acc_y","acc_z","or_x","or_y","or_z","grav_x","grav_y","grav_z"]

# # fix units
df.acc_x = df.acc_x/4096.0
df.acc_y = df.acc_y/4096.0
df.acc_z = df.acc_z/4096.0

df.rotR_x = df.rotR_x/16.4
df.rotR_y = df.rotR_y/16.4
df.rotR_z = df.rotR_z/16.4

df.grav_x = df.grav_x/4096.0
df.grav_y = df.grav_y/4096.0
df.grav_z = df.grav_z/4096.0
df = df.drop(columns="time")

#Basic features
df.insert(0, 'accel_x', df.acc_x + df.grav_x)
df.insert(1, 'accel_y', df.acc_y + df.grav_y)
df.insert(2, 'accel_z', df.acc_z + df.grav_z)
df.insert(3, 'accel_norm', np.sqrt(df.accel_x**2 + df.accel_y**2 + df.accel_z**2))
df = df.drop(['accel_x', 'accel_y', 'accel_z'], axis=1)



# A value is standardized as follows:

# y = (x – mean) / standard_deviation
# Where the mean is calculated as:

# mean = sum(x) / count(x)
# And the standard_deviation is calculated as:

# # standard_deviation = sqrt( sum( (x – mean)^2 ) / count(x))
# for col in df.columns:
#     mean = df[col].mean()
#     std = df[col].std()
#     df[col] = (df[col] - mean)/std


std = StandardScaler().set_output(transform='pandas')
df_std = std.fit_transform(df)
np.asarray(df).reshape(4,25,13)
```
